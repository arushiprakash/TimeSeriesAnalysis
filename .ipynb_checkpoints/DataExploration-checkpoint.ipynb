```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/prices.csv')
```

```python
data.columns
```

```python
stocks_of_interest = {
    'AMZN' : 'Amazon',
    'EBAY' : 'eBay'
}
```

```python
# Extracting the stocks of interest
amzn = data.loc[data['symbol'] == 'AMZN'] 
ebay = data.loc[data['symbol'] == 'EBAY'] 
```

## Visualizing the stock prices
It is important to visualize the data to get a feel of trend, seasonality and cyclic behaviour in the data.  
  
### Amazon Stock Visualization

```python
# Plotting the entire series
plt.plot(amzn['date'], amzn['close'], label='Amazon')
plt.xticks([])
plt.title("Closing stock prices over time")
plt.ylabel("Stock Price ($)")
plt.xlabel("Trading Day")
plt.legend()
plt.show()
```

```python
# Plotting a subset of data
sub_ = amzn.loc[(amzn['date'] < '2013-06-01') & (amzn['date'] > '2013-01-01')]
plt.plot(sub_['date'], sub_['close'], label='Amazon')
plt.xticks([])
plt.title("Closing stock prices over time")
plt.ylabel("Stock Price ($)")
plt.xlabel("Trading Day")
plt.legend()
plt.show()
```

**Observations** - Over the entire dataset, we see that the stock price has an increasing trend. There might also be some cyclic behavor every 1-1.5 years. On zooming into the data, a monthly seasonality is apparent.

## Seasonal Plots
We can also plot the yearly data against each other to see if there is some seasonality

```python
for yr in range(2010, 2017):
    sub_ = amzn.loc[pd.DatetimeIndex(amzn['date']).year == yr]
    plt.plot(pd.DatetimeIndex(sub_['date']).dayofyear, sub_['close'], label=yr)
plt.xticks()
plt.title("Closing stock prices over time")
plt.ylabel("Stock Price ($)")
plt.xlabel("Trading Day of Year")
plt.legend()
plt.show()
```

**Observations** - In most cases, there is an upward tick in stock prices towards the ends of the year and a slight dip in the 90-100 day mark (end of the first quarter).

## Lag Plots
Checking the correlation between different lagged values of the time series

```python
from pandas.plotting import lag_plot
lag_plot(amzn['close'], lag=1)
```

```python
lag_plot(amzn['close'], lag=2)
```

```python
lag_plot(amzn['close'], lag=5)
```

```python
lag_plot(amzn['close'], lag=50)
```

**Observation** - At lag = 1, the time series is ver correlated, which means that stock prices are correlated with the values of the previous day. As the lag increases, this correlation starts to break down.

## Autocorrelation Plots


```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(amzn['close'])
```

**Observations** - Since there is a trend in data, the autocorrelations at smaller lags are larger in size, and the ACF decreases over time. Time series characterised by slowly decaying ACF will usually exhibit a smooth pattern and can be classified as long-memory time series. There is little evidence for seasonal trends since when data are seasonal, the autocorrelations will be larger for the seasonal lags (at multiples of the seasonal frequency) than for other lags - a behavior that we don't see here.  
  
**Note** - or a white noise series, we expect 95% of the spikes in the ACF to lie within ±2/√T where T is the length of the time series.
