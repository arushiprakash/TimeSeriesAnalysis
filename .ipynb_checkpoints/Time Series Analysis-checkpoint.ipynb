# What is Time Series Analysis

Some key differences between a time series analysis and a standard analysis are laid out in the table below:
$$ $$

| . | Standard Analysis | Time Series Analysis|
---|---|---
**Does order matter?** | No | Yes
**How to model?** | Regression | Stochastic process
**How to characterize?** | Distribution | Distribution & Stationarity & ACF
**Judging model fit?** | Residuals | Characteristic similarity

$$ $$
Note that not every data set with a time component is necessarily a time series analysis.  The most important difference is whether the order of the data matters.  If it does (and the data are recorded over time) it's a time series analysis.

Two of the most important things to note in the comparison above are:
* The order of the points matters
* Time series are modeled with a stochastic process

The first point drives a lot of the interpretations of the data in time series analysis - the order represents something, so an FFT extracts meaningful structure.  Reordering the data destroys that structure (and can be done on a whim) which can be useful when trying to estimate various effects.  The second point is good to keep in mind during all the below discussions and demonstrations - the goal is to create a stochastic process which mimics all the important characteristics of the time series being analyzed.  Every observation, plot, and test calculation should be done with the question in mind: what process created this time series?

Broadly speaking, a stochastic process is one that is driven by a random noise input, but for time series analysis it tends to be structured as a process which has:
* Some degree of integration or differencing
* One or more possibly deterministic or exogenous processes
* One or more past values of the time series
* A random noise forcing
* One or more past values of the random noise forcing

The challenge of time series analysis can be understood as splitting the time series into the appropriate signals that capture the essence of the important effects at play in creating the time series.

### Plotting

One of the most important things you can do in any analysis is spend time visualizing your data.  Tables and statistics are good, but there is a certain impact of plots that can't be replaced by descriptive approaches.

There are three fundamental questions to ask:
* What rules do these data obey?
  * e.g. always > 0, integer-valued, variable time steps, categorical, etc.
* What patterns do these data have?
  * e.g. stationarity, seasonality, smoothness, periodicity, etc.
* What distribution do these data follow?
  * e.g. symmetry, modes, relation to parametric distributions, etc.

Below we have two statistically identical plots.  They have the same histogram, the same number of points, and the same mean and variance.  Do their time series plots look the same?

# Stationarity

A time series is said to be stationary if its statistical properties (mean, variance, autocorrelation) don't change over time.  Sometimes this is called "stationarity of order 2" or "second order stationary" by math nerds.  Another way to think about it is a time series is stationary if any contiguous subset of points are roughly representative of the overall series.

There are two main reasons for detecting and establishing stationarity in time series analysis:
* Models and statistical techniques like data that are stationary
* There are certain aspects of a nonstationary time series which may suggest the underlying process has changed - this is bad when trying to create a single stochastic process to characterize the time series and should be detected early.

Data scientists like data that are stationary.  Let's start with some tests for stationarity.  One test was already done - plotting.  Often visual inspection will decide whether a stationarity test needs to be run.  In this section there will be two kinds of stationarity addressed: mean stationarity and variance stationarity.

#### Mean Stationarity

The first component of a stationary time series is whether the mean changes with respect to time.  If the mean doesn't change the time series is stationary.  If it does change it's nonstationary.  Changing means wreak havoc on many statistical tests we use to identify useful inputs to models, and as such it should be understood and characterized in any time series analysis.

Another way to think of mean stationarity is it means the mean is not conditional on the time (forgive the abuse of notation):
$$ \mu(x|t_{[a,b]}) = \mu(x) $$

Where $\mu(x|t_{[a,b]})$ is the mean of the time series $x$ over the time range $[a,b]$ and $\mu(x)$ is the unconditional mean of $x$

A simple visual method to verify mean stationarity is calculating means and standard errors of the mean both for the overall series and for a window in the series.  Then plot the results.

```python
# Make a stationary and nonstationary process
y0 = np.random.randn(100)
y0[-25:] += np.arange(25)/6
y1 = np.random.randn(100)
y_set = [y0, y1]

# Calculate overall mean and SE_mean
m_set = []
se_set = []
for i0 in y_set:
    m_set.append(np.mean(i0))
    se_set.append(np.std(i0)/np.sqrt(len(i0)))

# Calculate windowed mean and SE_mean
window_len = 31
w_m_set = []
w_se_set = []
for i0 in y_set:
    c1 = np.convolve(i0, np.ones(window_len)/window_len, mode='valid')
    c2 = np.convolve(i0**2, np.ones(window_len)/window_len, mode='valid')
    std_window = np.sqrt(c2 - c1*c1)
    w_se_set.append(std_window / np.sqrt(window_len))
    w_m_set.append(c1)
    
# Visualize the result
c_i = 1.96
title0 = ['Nonstationary Mean', 'Stationary Mean']
for c0, (m0, se0, m1, se1) in enumerate(zip(m_set, se_set, w_m_set, w_se_set)):
    plt.subplot(1, len(y_set), c0+1)
    plt.plot([0, len(m1)], [m0, m0], linewidth=3, color=[0.4, 0.4, 0.4])
    plt.plot([0, len(m1)], [m0 - c_i*se0, m0 - c_i*se0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot([0, len(m1)], [m0 + c_i*se0, m0 + c_i*se0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot(m1, linewidth=3, color=[0.7, 0.3, 0.3])
    plt.plot(m1 + c_i*se1, 'r', linewidth=3)
    plt.plot(m1 - c_i*se1, 'r', linewidth=3)
    plt.title(title0[c0], fontsize=24)
plt.gcf().set_size_inches(12,5)
```

Why waste time plotting when a statistical test can tell you with more certainty?  Well, in time series analysis it's important not just whether something happens, but when it happens.  One of the decisions that can be made while processing time series is whether to split them into smaller contiguous chunks which exhibit more consistent behavior.  This is related to a breakpoint model or regime changing model.

Our nonstationary time series here has a characteristic hockey stick appearance which may indicate it needs to be transformed or there may have been an important exogenous event at the inflection point that will need to be investigated.  Either way there is valuable information in the plot that isn't present in a statistical test and p-value.

#### Variance Stationary

The second component in stationarity is the variance, which can also be observed as the spread of data as a function of time.  If a time series has changing variance with respect to time it is called heteroskedasticity and the time series is not stationary.

$$ \sigma^{2}_{x|t_r} = \sigma^{2}_{x} $$

Where $\sigma^{2}_{x|t_r}$ is the variance of the time series $x$ over the time range $t_r$ and $\sigma^{2}_{x}$ is the unconditional variance of $x$.  Also, note this is even more of an abuse of notation and definitions than for the mean, as, in general, we wouldn't expect the variance to be exactly equal, but, more likely, that all the sample variances act as if they're drawn from the same distribution.

Again it is possible to get a qualitative feel for the degree and nature of any nonstationarity by plotting the variance in a window as that window moves across the time series.

```python
def k_n(n):
    """
    Numerically stable term to correct for bias in the estimate of standard 
    deviation, K_n, used to construct the UMVU estimate of standard deviation:
    
    K_n * sample_std => unbiased sample std
    
    Uses a Stirling approximation for Gamma to handle larger(ish) numbers.
    https://web.eecs.umich.edu/~fessler/papers/files/tr/stderr.pdf
    """
    n = (n-1)/2
    p0 = (n-0.5)*np.log(n) - n*np.log(n+0.5) + 0.5
    return(np.sqrt(n)*np.exp(p0))
```

```python
# Make a stationary and nonstationary process
y0 = np.random.randn(500)*(1+np.arange(1,501)/100)
y1 = np.random.randn(500)
y_set = [y0, y1]

# Calculate overall standard deviation and SE_stdev
sd_set = []
se_set = []
for i0 in y_set:
    s0 = np.std(i0)*k_n(len(i0))
    if len(i0) > 340:
        v0 = 0.5
    else:
        v0 = chi.stats(len(i0)-1, moments='v')
    sd_set.append(s0)
    se_set.append(s0*np.sqrt(v0/(len(i0)-1)))

# Calculate windowed standard deviation and SE_stdev
window_len = 101
w_sd_set = []
w_se_set = []
for i0 in y_set:
    c1 = np.convolve(i0, np.ones(window_len)/window_len, mode='valid')
    c2 = np.convolve(i0**2, np.ones(window_len)/window_len, mode='valid')
    std_window = np.sqrt(c2 - c1*c1)
    v1 = k_n(window_len)*np.sqrt((chi.stats(window_len-1, moments='v'))/(window_len-1))
    w_se_set.append(std_window * v1)
    w_sd_set.append(std_window)
    
# Visualize the result
c_i = 1.96
title0 = ['Nonstationary StDev', 'Stationary StDev']
for c0, (m0, se0, m1, se1) in enumerate(zip(sd_set, se_set, w_sd_set, w_se_set)):
    plt.subplot(1, len(y_set), c0+1)
    plt.plot([0, len(m1)], [m0, m0], linewidth=3, color=[0.4, 0.4, 0.4])
    plt.plot([0, len(m1)], [m0 - c_i*se0, m0 - c_i*se0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot([0, len(m1)], [m0 + c_i*se0, m0 + c_i*se0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot(m1, linewidth=3, color=[0.7, 0.3, 0.3])
    plt.plot(m1 + c_i*se1, 'r', linewidth=3)
    plt.plot(m1 - c_i*se1, 'r', linewidth=3)
    plt.title(title0[c0], fontsize=24)
plt.gcf().set_size_inches(12,5)
```

Keeping in mind that plots are powerful tools for figuring out *how* a time series is nonstationary, it's still helpful to have a statistical test to assign a numerical significance to and simple characterization of the heteroskedasticity

One of the most well known tests for heteroskedasticity is the Brusch-Pagan test.

#### Stationary Autocorrelation

Autocorrelation measures the strength of the linear relationship between a time series and itself, lagged some number of points.  Taken across all the lags, this forms the autocorrelation function.  

The autocorrelation function represents how well a time series predicts itself, but the results are hard to interpret for time series that exhibit a trend.  Assuming a time series is mean stationary, higher autocorrelation values at longer time periods represent a longer-term dependence on the past values of the series.  If a trend exists in the data it can look like a strong dependence at a long time lag.  In fact, an ACF that doesn't asymptote down to zero is evidence of a trend in the data.

In addition to the ACF, there is also the partial autocorrelation function.  The correlation at each lag has the effect from any previous lags removed.  This is done in a manner similar to gram-schmidt orthogonalization, removing the orthogonal projection to the subspace spanned by earlier lags in the time series.  

In ARIMA models, the number of lags with statistically significant corrrelation indicate the order of the AR part of the model.  The number of lags with statistically significant correlation in the PACF indicate the order of the MA part of the model. 

```python
# Generate two series
y0 = np.random.randn(1000)
y1 = np.convolve(np.random.randn(1000), np.ones(6)/6, mode='same')
y0[500:] = np.convolve(y0[500:], np.ones(6)/6, mode='same')
y_set = [y0, y1]

# Calculate the overall autocorrelation and uncertainty window
window_len = 101
tau = 1
ac0 = []
ac0_stderr = []
for i0 in y_set:
    ac0.append(np.sum(i0[tau:]*i0[:-tau])/np.sum(i0*i0))
    ac0_stderr.append(1.96/(len(i0)-tau))

# Calculate autocorrelation and uncertainty in a window
ac1 = []
ac1_stderr = []
for i0 in y_set:
    covar0 = np.convolve(i0[tau:]*i0[:-tau], np.ones(window_len), mode='valid')
    var0 = np.convolve(i0*i0, np.ones(window_len+1), mode='valid')
    ac1.append(covar0/var0)
    ac1_stderr.append(1.96/np.sqrt(window_len-tau))

# Plot
title0 = ['Nonstationary Autocorr. (lag %d)'%(tau), 
          'Stationary Autocorr. (lag %d)'%(tau)]
for c0, (i0, e0, i1, e1, t0) in enumerate(zip(ac0, ac0_stderr, ac1, ac1_stderr, title0)):
    plt.subplot(1,2,c0+1)
    upper1, lower1 = i1+e1, i1-e1
    upper1[upper1 > 1] = 1
    upper1[upper1 < -1] = -1
    lower1[lower1 > 1] = 1
    lower1[lower1 < -1] = -1
    plt.plot([0, len(y0)-window_len], [i0, i0], linewidth=3, color=[0.4, 0.4, 0.4])
    plt.plot([0, len(y0)-window_len], [i0-e0, i0-e0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot([0, len(y0)-window_len], [i0+e0, i0+e0], linewidth=3, color=[0.7, 0.7, 0.7])
    plt.plot(i1, linewidth=3, color=[0.7, 0.3, 0.3])
    plt.plot(lower1, 'r', linewidth=3)
    plt.plot(upper1, 'r', linewidth=3)
    plt.title(t0, fontsize=24)
plt.gcf().set_size_inches(12,5)
```

One test for autocorrelation is the Durbin-Watson test, which is usually used to test for autocorrelation in the residuals of a model, but also works just as well on any discrete time series or subsequence of a time series.  The estimates and uncertainties can characterize the stationarity of the autocorrelation.

The main question asked during TSA is: what stochastic process represents this time series?  In the course of answering that question, several other questions arise:
* What rules does my time series obey?
* Is my time series stationary (second order stationary, generally)?
  * If not, consider differencing or detrending the series.
* Is my time series constructed of only one signal?
  * Does it look like there are time steps where the underlying process changes?
    * If so, consider breaking the time series into homogeneous sub-series
  * What is (are) the distribution(s) of this time series?
    * Consider transforms to change domain, symmetry, etc.
  * Does it look like there are multiple overlapping signals in the time series?
    * If so, consider decomposing the signal (covered below)
    * Note this also covers things like ACF, PACF, and FFT plots which weren't covered in this first part of stationarity, but will be covered to some degree later.
 
These are the initial observations that will drive the second part of time series analysis: processing the time series.  This is where the time series is broken down into pieces, each piece is analyzed, and the whole picture of what's driving the (stochastic) time evolution of the process is formed and tested.

## Adversarial Example

Establishing stationarity can be a very comforting thing in a time series analysis.  Create a few plots or run a few statistical tests and, if the time series passes, the model building can start with full confidence.

Except it can't.

While the definition of stationarity here (second order stationarity) is well established in the literature and is sufficient for most time series, it's still possible to construct an adverserial example which will, by construction, pass every test and still clearly fail to be a time series constructed from a single stochastic process.

```python
# Create a time series
r0 = np.random.randn(500)+1
r1 = np.random.exponential(size=500)
r1 = r1 / np.std(r1)
r1 = r1 - np.mean(r1) + 1
r2 = np.concatenate([r0, r1])

# Calculate overall mean and SE_mean
m_set = np.mean(r2)
se_set = np.std(r2)/np.sqrt(len(r2))

# Calculate windowed mean and SE_mean
window_len = 201
c1 = np.convolve(r2, np.ones(window_len)/window_len, mode='valid')
c2 = np.convolve(r2**2, np.ones(window_len)/window_len, mode='valid')
std_window = np.sqrt(c2 - c1*c1)
w_se_set = std_window / np.sqrt(window_len)
w_m_set = c1

# Calculate overall standard deviation and SE_stdev
s0 = np.std(r2)*k_n(len(r2))
if len(r2) > 340:
    v0 = 0.5
else:
    v0 = chi.stats(len(r2)-1, moments='v')
sd_set = s0
se_set = s0*np.sqrt(v0/(len(r2)-1))

# Calculate windowed standard deviation and SE_stdev
c1 = np.convolve(r2, np.ones(window_len)/window_len, mode='valid')
c2 = np.convolve(r2**2, np.ones(window_len)/window_len, mode='valid')
std_window = np.sqrt(c2 - c1*c1)
v1 = k_n(window_len)*np.sqrt((chi.stats(window_len-1, moments='v'))/(window_len-1))
w_se_set = std_window * v1
w_sd_set = std_window

# Calculate the overall autocorrelation and uncertainty window
tau = 1
ac0 = np.sum(r2[tau:]*r2[:-tau])/np.sum(r2*r2)
ac0_stderr = 1.96/(len(r2)-tau)

# Calculate autocorrelation and uncertainty in a window
covar0 = np.convolve(r2[tau:]*r2[:-tau], np.ones(window_len), mode='valid')
var0 = np.convolve(r2*r2, np.ones(window_len+1), mode='valid')
ac1 = covar0/var0
ac1_stderr = 1.96/np.sqrt(window_len-tau)
    
# Visualize the result
c_i = 1.96
plt.subplot(4,1,1)
plt.plot([0, len(w_m_set)], [m_set, m_set], linewidth=3, color=[0.4, 0.4, 0.4])
plt.plot([0, len(w_m_set)], [m_set - c_i*se_set, m_set - c_i*se_set], 
         linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot([0, len(w_m_set)], [m_set + c_i*se_set, m_set + c_i*se_set], 
         linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot(w_m_set, linewidth=3, color=[0.7, 0.3, 0.3])
plt.plot(w_m_set + c_i*w_se_set, 'r', linewidth=3)
plt.plot(w_m_set - c_i*w_se_set, 'r', linewidth=3)
plt.title("Totally Reliable Plot of Windowed Mean", fontsize=24)

# Visualize the result
plt.subplot(4,1,2)
plt.plot([0, len(w_sd_set)], [sd_set, sd_set], linewidth=3, color=[0.4, 0.4, 0.4])
plt.plot([0, len(w_sd_set)], [sd_set - c_i*se_set, sd_set - c_i*se_set], 
         linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot([0, len(w_sd_set)], [sd_set + c_i*se_set, sd_set + c_i*se_set], 
         linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot(w_sd_set, linewidth=3, color=[0.7, 0.3, 0.3])
plt.plot(w_sd_set + c_i*w_se_set, 'r', linewidth=3)
plt.plot(w_sd_set - c_i*w_se_set, 'r', linewidth=3)
plt.title("Totally Reliable Plot of Windowed Standard Deviation", fontsize=24)

# Plot
plt.subplot(4,1,3)
upper1, lower1 = ac1+ac1_stderr, ac1-ac1_stderr
upper1[upper1 > 1] = 1
upper1[upper1 < -1] = -1
lower1[lower1 > 1] = 1
lower1[lower1 < -1] = -1
plt.plot([0, len(ac1)], [ac0, ac0], linewidth=3, color=[0.4, 0.4, 0.4])
plt.plot([0, len(ac1)], [ac0-ac0_stderr, ac0-ac0_stderr], linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot([0, len(ac1)], [ac0+ac0_stderr, ac0+ac0_stderr], linewidth=3, color=[0.7, 0.7, 0.7])
plt.plot(ac1, linewidth=3, color=[0.7, 0.3, 0.3])
plt.plot(lower1, 'r', linewidth=3)
plt.plot(upper1, 'r', linewidth=3)
plt.title("Totally Reliable Plot of Windowed Autocorrelation", fontsize=24)

plt.subplot(4,1,4)
plt.plot(r2)
plt.title("Totally Normal Stationary Time Series", fontsize=24)
plt.gcf().set_size_inches(12,26)

print(durbin_watson(r2))
print(durbin_watson(r2-np.mean(r2)))
```

All of the visualizations failed.  Or, rather, they didn't fail, but it's important to be aware of the definitions and assumptions being used, here.

Advanced statistical tests wouldn't help here, either.  Everything that has been done up to this point is to establish second order stationarity.  If the mean, the variance, and the autocorrelation aren't dependent on time, the time series is second order stationary.  The time series generated here is clearly second order stationary and should, therefore, pass all the tests with confidence-inspiring significance.  However, this time series clearly isn't generated by a single process, and would perform very poorly with most model fits.

It's worthwhile getting comfortable with the estimation of these parameters and their uncertainty, as well as visualizing time series with different rules, different distributions, and different levels of stationarity.

# Statistical Tests

These are many available tests which make various assumptions about the time series being modeled and, within the limits of validity of those assumptions, allow a well-founded and objective judgment of the stationarity of the time series.

Here we will cover a sample of three of these tests which are commonly used in characterizing time series.  The important thing to understand for any of these tests is which assumptions they make about the time series (e.g. modeled well by AR(1), normally distributed, etc.) and to make an independent judgment of how well the time series of interest meets those assumptions.



#### Augmented Dickey-Fuller Test

The Augmented Dickey-Fuller test tries to detect if a time series is characterised by a unit root process.  It does this by fitting a model to the time series and tests the signficance of the coefficients.

In one case, if the time series is stationary it will revert to some fixed mean.  This means that the actual value of the time series in the previous time series should have an effect on how the time series evolves, since being further from the mean will tend to push the process back towards the long term mean.

If the time series is nonstationary it won't have a tendency to move back toward any particular mean, and there won't be any predictive power from the value of the time series in the previous time step.

Mathematically, the way I like to think about unit roots is in the context of the backshift operator, commonly denoted $B$.  The backshift operator works by stepping back one time increment in a time series:

$$ B(x_t) = x_{t-1} $$
$$ B^2(x_t) = x_{t-2} $$

This permits writing an autoregressive relationship like a polynomial equation, for example:

$$x_t = a_0 + a_1B(x_t) + a_2B^2(x_t) + a_3B^3(x_t) + ... + a_nB^n(x_t)$$

The roots of the polynomial can then be solved for in the traditional methods:

$$(B(x)-c_0)(B(x)-c_1)(B(x)-c_2)...(B(x)-c_n)$$

If any of the values $c_0 ... c_n$ are equal to one, then the root (literally a value where the polynomial equals 0) is a unit root, and the process will be nonstationary.

An equivalent formulation of the ADF test is to represent the process as an AR(1) like this:

$$ \Delta y_t = c_0 + c_1t + \gamma y_{t-1} + ... + \epsilon_t$$

In this formulation the test is whether $\gamma = 0$ which can be interpreted as saying that the current level of the time series has no impact on the next change.  This is a characteristic of time series with a unit root - shocks to the system are permanent and the time series shows no tendency to move back toward a particular long term trend.  If $\gamma \neq 0$ then the level of the time series determines the change, that is, if the time series get too high or too low it will revert back to the mean.

The ADF test is flexible enough to allow a constant and a time-dependent trend in the time series, but this flexibility is also a difficulty.  Because determining the existence of signals like trends is a core piece of time series analysis there's often no way to know, beforehand whether a constant or linear trend should be applied to the change in $y$.

As a consequence of this, if there should be, say, a linear trend term and it's missing, it will rob the ADF of its statistical power for rejecting the null hypothesis (for unit roots).  In contrast if the linear trend term is included in ADF when it shouldn't be there then the test is misspecified which may invalidate the result.

The general procedure for performing the ADF on a time series with unknown signals roughly follows these steps:
1. Run ADF with trend and constant
  * If the null hypothesis is rejected, the time series is stationary
  * If a unit root might be present, verify the trend with independent testing.  If the trend is verified, assume a unit root is present.  Otherwise move on to step 2
2. Re-run the ADF with just the constant.
  * If the null hypothesis is rejected, the time series is stationary
  * If a unit root might be present, verify the constant term with independent testing.  If the constant term is verified, assume a unit root is present.  Otherise move onto step 3
3. Re-run the ADF with no constant or trend
  * If the null hypothesis is rejected, the time series is stationary, otherwise there is a unit root

Independent verification would look something like running a parameter estimation on something like:

$$ \Delta y_t = c_0 + c_1t + \Sigma \beta_i \Delta y_{t-i} $$

And evaluating the significance of the coefficients.  This formulation has no regressors for the integration of order 1, so it can be used to test for the presence of a significant $c_0$ or $c_1$ using a t-test

```python
y0 = np.random.randn(1000)
y1 = np.cumsum(y0)

res0 = adfuller(y0, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False)
res1 = adfuller(y1, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False)

print(f"ADF p-value for stationary series {res0[1]}")
print(f"ADF p-value for unit root series {res1[1]}")
```

#### Durbin Watson

The Durbin Watson test is designed to test for autocorrelation in the residuals of a model fit - since the presence of autocorrelation in that instance would indicate something missing or misspecified for the model.  For time series analysis there's an expectation of autocorrelation in the signal, but it's useful to have a lightweight test to quantify it at each stage of signal decomposition.

The Durbin Watson test is just that lightweight reference point.  It doesn't have a strict definition for statistical significance or a strong threshold between "has autocorrelation" and "doesn't have autocorrelation."  The test returns a value betwen 0 and 4, with a value of 2 indicating "no autocorrelation."  If the value is between 0 and 2 it indicates some amount of positive autocorrelation.  Between 2 and 4 is some amount of negative autocorrelation.  Generally values between 1.5 and 2.5 are accepted as "neutral" or "no autocorrelation."

The test itself is relatively easy to compute and understand.  It's the sum of the variance of the difference of the time series divided by the variance of the time series.

$$ \frac{\Sigma_{i=2}^{n} (x_i-x_{i-1})^2}{\Sigma_{i=1}^{n} x_i^2} $$

Where $x_i$ is the $i^th$ observation in the time series, and $n$ is the length of the time series.  The expected result for zero autocorrelation then says that the variance of the difference of sequential points will be twice the normal variance.  This only makes sense if the results are 0 mean (because there are chances of that difference making the result larger or smaller when the $x_i$ are 0 mean, while the result will always be smaller if the $x_i$ are, say, strictly positive), which is guaranteed for most model residuals, but not for arbitrary time series.  Therefore, when using the Durbin Watson test, be sure to remove the mean of the time series before evaluating the Durbin Watson.

```python
y0 = np.random.randn(1000)
y1 = np.convolve(np.random.randn(1000), np.ones(7)/7, mode='same')
y1 = y1 / np.std(y1)

print(f"Durbin Watson for IID data            {np.round(durbin_watson(y0),2)}")
print(f"Durbin Watson for autocorrelated data {np.round(durbin_watson(y1),2)}")

plt.plot(y0, label='IID', color=[0.2,0.2,0.5])
plt.plot(y1, label='Autocorrelated', color=[0.5,0.5,0.2])
plt.legend()
plt.gcf().set_size_inches(12,5)
```

# Decomposition

A more nuanced approach to inducing stationarity into a time series involves decomposing the signal, usually into three or four pieces:

* Trend
* Cycle
* Season
* Residual

Often the trend and cycle are combined into a single signal.  There are two main types of decomposition: additive and multiplicative.  In additive decomposition each signal is added together to reconstruct the original time series.  In multiplicative decomposition each signal is multiplied together to reconstruct the original time series.  

While multiplicative decomposition is good for explaining systems where the variance looks like it depends on the level (mean) of the signal - that is, the mean and variance are both nonstationary, but look to be linearly related to each other - I find in practice that since the multiplicative model has an equal contribution from each component, it can be difficult to control the noisiness of the resulting time series.  In other words, because all the components are multiplied, if your random noise is 0.8, it reduces the point in the time series (trend and seasonality) by 20%, while a random noise of 0.8 in an additive model is likely to be negligable.  It all depends on the time series in question, but be aware of the sensitivity of multiplicative decomposition models

### Trend

Three main trend methods will be explored:

* Linear fit
* Filter / FFT
* LOWESS

There are several overlapping goals for trend identification and removal: first, it can be a useful tool for creating a mean stationary time series, second it can characterize low frequency trends, and third it can reveal cyclical components that aren't clear in the noisy original time series.

#### Linear Fit

This approach assumes the trend can be approximated by an OLS fitted line.

Advantages:
* Simple
* Hard to overfit
* No edge effects

Disadvantages:
* Doesn't respect rules of the time series (e.g. >= 0)

#### Filter / FFT

These can be thought of as largely frequency-based approach.  The FFT is a key tool in analyzing and understanding time series.  There are two FFT-based methods showcased here: the first one converts the time series to frequency space and then sets all values above a cutoff frequency to 0.  This can be an attractive approach because frequency thresholds can be chosen based on the understanding of the time series (e.g. picking a threshold below the known seasonal frequency).  This effectively constructs a trend out of the lowest-frequency components in the time series, but it suffers from several problems:
* FFTs on nonstationary data can be tricky to interpret (although not **technically** incorrect to do)
* Heavy-handedly setting values to 0 (using any "hard" filter creates distortions when switching between frequency domain and time domain - e.g. boxcar filter v gaussian filter)

A slightly more nuanced way to implement this idea is with a Butterworth (or other) filter.  The Butterworth filter is interesting in that it produces a maximally flat response in the time domain, given the fastest rate of attenuation in the frequency space.  This will "blend out" the frequency components higher than the cutoff point rather than setting them to 0.  This should provide the same benefits of an FFT approach, but will suffer from less stark edge effects and ripples in the time domain.

A filter approach is closely related to the FFT approach, and only really varies in the details of the implementation.  An arbitary finite impulse response filter can be constructed to create predetermined changes to the frequency space.  For the filter approach this frequency modification is indirectly accomplished through a filter, with different parameter choices: filter shape and filter length.  

#### LOWESS

LOWESS (or LOESS, LOocally WEighted Scatterplot Smoothing) is a nonparametric smoother which locally fits a low order polynomial (often a line) to a subset of the data in a particular window.  This window is moved over the time series and at regular intervals the low order polynomial is estimated again.  This creates a bunch of local, low order polynomial approximations of the data.  These are all blended together by assuming the low order polynomial fit is most relevant in the center of the window, and less relevant further and further out.

The blended low order polynomial becomes a nonparametric estimate of the smoothed time series.  The parameter used to control the fit is the width of the window, sometimes noted as the "fraction of points" included in the window.  The lower this fraction of points the smaller the window and the more tightly the LOWESS estimator will follow the time series.  I like to use values between 25% (0.25) and 50% (0.5) for trend estimation in time series analysis, but other values are reasonable depending on the time series.

LOWESS, like the FFT / filter approach, suffers from edge effects and that can make it more difficult to use in a forecasting context because it is least accurate where the trend is needed most - at the leading edge of the time series.

#### Statsmodels

I've also included the fit from statsmodels seasonal_decompose function which estimates the trend, seasonality, and residual for a time series.

As a general note, all of these approaches are very sensitive to the parameter choices made.  It can be hard to objectively be certain in a set of parameter choices, and a lot of decomposition comes down to how a series looks, feels, and what the data scientist believes is the structure of the underlying signals.  Sometimes domain knowledge of the time series can be helpful in narrowing down options for decomposition.

```python
# Time series to decompose
y_sz = 1000
acf_filt = np.random.rand(5)
acf_filt = acf_filt / np.sum(acf_filt)
acf_filt = np.arange(5)+1
acf_filt = acf_filt/np.sum(acf_filt)
y0 = np.convolve(3*np.random.randn(y_sz), acf_filt, mode='same')
y0 += 1.5*np.cos(np.linspace(0,20*np.pi, y_sz)) 
y0 += np.sin(np.linspace(0,10*np.pi, y_sz))
y0 += np.sin(np.linspace(0,40*np.pi, y_sz))
y0 += 5e-8*np.arange(-500,500)**3 

# Linear fit
# two parameters - intercept and slope
fit0 = np.polyfit(np.arange(1000), y0, deg=1)

# FFT Fit
# one parameter - frequency to keep
# Multiple parameters if doing optimal rolloff (e.g. butterworth filter)
y0_f = np.fft.fft(y0)
freq = np.fft.fftfreq(len(y0))
# FFT for hard removal of frequencies
keep_freq = 0.004
zero_ind = np.where(np.abs(freq) > keep_freq)
y0_f[zero_ind] = 0 
fit1 = np.fft.ifft(y0_f)
# Butterworth for optimally smooth frequency removal
keep_freq = 0.008
b, a = signal.butter(3, keep_freq, 'low', analog = False)
fit1a = signal.filtfilt(b, a, y0).real

# Filter Fit
# One parameter - filter length
# Multiple parameters if shaping the filter (e.g. linear falloff)
f_sz = 200
fit2 = np.convolve(y0, np.ones(f_sz+1)/(f_sz+1), mode='same')
fit2a = np.convolve(y0, np.ones(-20+f_sz+1)/(-20+f_sz+1), mode='same')

# "Official" Fit
# Single parameter - frequency of seasonality
fit3 = seasonal_decompose(y0, freq=200)

# LOWESS Fit
fit4 = lowess(y0, np.arange(len(y0)), frac=0.35)

# Detrended data
y0_nt = y0 - fit4[:,1]

# Plot results
plt.subplot(6,1,1)
plt.plot(y0)
plt.plot(fit0[1] + np.arange(1000)*fit0[0], linewidth=3, label='Linear')
plt.plot(5e-8*np.arange(-500,500)**3, linewidth=3, label='Actual')
plt.legend()
plt.title("Time Series with Linear Trend Fit", fontsize=24)
plt.subplot(6,1,2)
plt.plot(y0)
plt.plot(fit1, linewidth=3, label='FFT')
plt.plot(fit1a, linewidth=3, label='FFT_Butter')
plt.plot(5e-8*np.arange(-500,500)**3, linewidth=3, label='Actual')
plt.legend()
plt.title("Time Series with FFT Trend Fit", fontsize=24)
plt.subplot(6,1,3)
plt.plot(y0)
plt.plot(fit2, linewidth=3, label='Filter')
plt.plot(5e-8*np.arange(-500,500)**3, linewidth=3, label='Actual')
plt.legend()
plt.title("Time Series with Filter Trend Fit", fontsize=24)
plt.subplot(6,1,4)
plt.plot(y0)
plt.plot(fit3.trend, linewidth=3, label='Statsmodels')
plt.plot(5e-8*np.arange(-500,500)**3, linewidth=3, label='Actual')
plt.legend()
plt.title("Time Series with Statsmodels Trend Fit", fontsize=24)
plt.subplot(6,1,5)
plt.plot(y0)
plt.plot(fit4[:,1], linewidth=3, label='LOWESS')
plt.plot(5e-8*np.arange(-500,500)**3, linewidth=3, label='Actual')
plt.legend()
plt.title("Time Series with LOWESS Trend Fit", fontsize=24)
plt.subplot(6,1,6)
plt.plot(y0_nt)
plt.title("Time Series with Trend Removed (LOWESS)", fontsize=24)

plt.gcf().set_size_inches(12,36)
```

### Seasonality

After the trend (trend + cycle) has been removed, the next component to address is seasonality.  This is cyclical signal with a fixed period that occurs multiple times in the sample observation period.

Note the the seasonality signal itself may be a complex shape - it's not important whether the seasonality is sinusoidal or parametric, just that the pattern is repeated at regular intervals in the detrended data.

We'll be covering several methods of estimating seasonality:
* Periodic functions
* Filter / FFT
* Periodic means
  * Changing Magnitudes

#### Periodic Functions

There are any number of periodic functions (e.g. sin, cos) which can be used to capture the essential nature of the periodic component in a time series.  Sometimes the choice is based empirically off of observations, or sometimes it's because of domain knowledge of the signal (e.g. a time series of A/C power should show a 60hz sine wave in the US).  If a periodic function is chosen it will often have a theoretical justification associated with it: why this particular function would be expected to arise in the time series.

#### FFT

Much in the same way that the lowest frequencies can be pulled out at a certain frequency threshold to estimate the trend, a second bucket of low frequency signal can be pulled out for seasonality.  This has all the advantages (convenience, reasonable justification for frequency choice) and disadvantages (edge effects, filter characteristics) as the FFT approach for trend isolation.

#### Periodic Means

In my opinion this is the gold standard for removing seasonality with a known period.  The time series data are reshaped into a matrix where each row represents an instance of the seasonality (rows have a length of D, the period of the seasonality).  The columns are then averaged to generate an "average seasonality" which can then be isolated from the rest of the signals.

It's common for the size of the seasonality to change over the course of a time series.  This can be seen in the residuals after removing the seasonality.  An inverse seasonal signal will occur in the earlier part of the time series, which will fade to nothing in the middle, and then emerge as a positive seasonal signal in the later part of the time series.  If this happens there are a few things to try:
* Change to a multiplicative model
  * At the cost of increased sensitivity to residuals, the seasonality may scale with the level of the time series and this is a natural way to incorporate that aspect of the signal
* Rescale each seasonal value
  * This introduces extra parameters and complicates forecasting, but can create the cleanest residual signal as each season is best fit to the magnitude of the seasonal signal.

```python
# Periodic function
# Not just constrained to sin/cos - anything periodic that resembles the time series
# Two parameters, estimate magnitude and frequency
best_score, best_period = -1, None
for i0 in range(1,50):
    sig0 = np.cos(np.linspace(0,i0*2*np.pi,len(y0_nt)))
    new_score = np.corrcoef(sig0, y0_nt)[0,1]
    if np.abs(new_score) > best_score:
        best_score = new_score
        best_period = i0
wt0 = np.median(y0_nt / np.cos(np.linspace(0,best_period*2*np.pi,len(y0_nt))))
fit0 = wt0*np.cos(np.linspace(0,best_period*2*np.pi,len(y0_nt)))

# FFT / Filter
y0_f = np.fft.fft(y0_nt)
freq = np.fft.fftfreq(len(y0_nt))
# FFT for hard removal of frequencies
keep_freq = 0.08
zero_ind = np.where(np.abs(freq) > keep_freq)
y0_f[zero_ind] = 0 
fit1 = np.fft.ifft(y0_f)
# Butterworth for optimally smooth frequency removal
# Technically this could be a band pass, but I've had more
# success removing the trend and structuring the filter
# as a low pass
keep_freq = 0.08
b, a = signal.butter(3, keep_freq, 'low', analog = False)
fit1a = signal.filtfilt(b, a, y0_nt).real
# Extra required step - standardize seasonality
fit1 = np.tile(np.mean(fit1.reshape((5,200)),axis=0).ravel(), 5)
fit1a = np.tile(np.mean(fit1a.reshape((5,200)),axis=0).ravel(), 5)

# Periodic means
fit4 = np.tile(np.mean(y0_nt.reshape((5,200)), axis=0).ravel(), 5)

# Residual
y0_r = y0_nt - fit4

# Plot results
plt.subplot(6,1,1)
plt.plot(y0_nt)
plt.plot(fit0 , linewidth=3, label='Periodic')
plt.legend()
plt.title("Time Series with Periodic Function Seasonality Fit", fontsize=24)
plt.subplot(6,1,2)
plt.plot(y0_nt)
plt.plot(fit1, linewidth=3, label='FFT')
plt.plot(fit1a, linewidth=3, label='FFT_Butter')
plt.legend()
plt.title("Time Series with FFT Seasonality Fit", fontsize=24)
plt.subplot(6,1,3)
plt.plot(y0_nt)
plt.plot(fit3.seasonal, linewidth=3, label='Statsmodels')
plt.legend()
plt.title("Time Series with Statsmodels Seasonality Fit", fontsize=24)
plt.subplot(6,1,4)
plt.plot(y0_nt)
plt.plot(fit4, linewidth=3, label='Periodic Mean')
plt.legend()
plt.title("Time Series with Periodic Mean Seasonality Fit", fontsize=24)
plt.subplot(6,1,5)
plt.plot(y0_nt - fit4)
plt.title("Time Series with Period Removed (Periodic Mean)", fontsize=24)
plt.subplot(6,1,6)
plt.plot(fit0[:200], linewidth=3, label='Periodic Function')
plt.plot(fit1[:200], linewidth=3, label='FFT')
plt.plot(fit1a[:200], linewidth=3, label='FFT Butter')
plt.plot(fit3.seasonal[:200], linewidth=3, label='Statsmodels')
plt.plot(fit4[:200], linewidth=3, label='Periodic Mean')
plt.legend()

plt.gcf().set_size_inches(12,36)

# plt.subplot(2,1,1)
# plt.plot(y0_nt)
# plt.plot(1.54*np.cos(np.linspace(0,best_period*2*np.pi,len(y0_nt))))
# plt.subplot(2,1,2)
# plt.plot(y0_nt-1.54*np.cos(np.linspace(0,best_period*2*np.pi,len(y0_nt))))
```

### Residual

Once the residual is reached the concern is mostly around "does this residual act like a good residual?" and is often used to judge the quality of the trend and seasonality removal.  If the residuals don't look good it's a sign to check the assumptions that went into the trend or seasonality removal and adjust accordingly.

That means stationarity tests and plots should be run on the residuals to ensure they behave in an acceptable way.  In this example (with a cubic trend) the time series is has exaggerated edge effects which affect the residuals.  In a full analysis this may have to be dealt with.

It's common for time series residuals to exhibit autocorrelation, but it can be useful to have uncorrelated residuals.  A way to approach this is to use a Wiener filter on the ACF to design a filter which will transform white noise into the characteristic noise of the residual.

The general steps for a Wiener filter are slightly involved and include solving a system of equations relating matrices of the autocorrelation functions to each other (through convenient Toeplitz matrices, but still a bother).  Fortunately since our input signal is statistical white noise, the matrix of ACF becomes the identity matrix.  Easy to invert and easy to solve equations with.  The end result is that the filter which enforces the target autocorrelation onto white noise is, identically, the ACF of the target autocorrelation.

Due to how np.convolve works with filters in Python, be sure to reverse the order of the ACF before filtering with it.

```python
filt0 = tsa.stattools.acf(y0_r, nlags=20)[::-1]
y_sim = np.convolve(np.random.randn(1000), filt0, mode='same')

plt.subplot(3,1,1)
plt.plot(tsa.stattools.acf(y0_r, nlags=20), label='Actual')
plt.plot(tsa.stattools.acf(y_sim, nlags=20), label='Simulated')
plt.title("ACF for Filtered White Noise")
plt.legend()
plt.subplot(3,1,2)
plt.plot(y0_r, label='Actual')
plt.plot(np.random.randn(1000)-6, label='White noise')
plt.legend()
plt.subplot(3,1,3)
plt.plot(y0_r, label='Actual')
plt.plot(y_sim-6, label='Simulated' )
plt.legend()
plt.gcf().set_size_inches(12,18)
```

# Differencing

Another approach to establish stationarity in a time series is to difference it - that is, take the discrete equivalent of the first derivative.  Functionally this means taking the difference (or in some cases, ratio) of sequential points in the time series.

This approach is useful for time series with a unit root, as those will not show a preference for any trend (although they may appear to follow a trend) and, therefore, (**unfounded opinion**) decomposition would be misleading at best.

If a time series has a unit root that implies that, while the time series itself isn't stationary, the first difference of the time series is stationary.  The order of integration, though, isn't a binary value that can only be 1 or 0.  Fractionally integrated time series demonstrate characteristic behavior that puts them between stationary time series and time series integrated of order 1.

A time series integrated of order less than one which is differenced can show "overdifferencing" - generally characterized by negative autocorrelations and more rapidly varying values.

Fractionally integrated time series show long term dependence and slowly decaying frequency spectra which require a different class of models to correctly capture their behavior (fractional Gaussian noise), therefore identifying the correct order of integration is key to properly modeling and differencing a time series.

Time series integrated of order more than 1 are very rarely encountered in time series based on observational, real world data.  A time series integrated of order 1 can have a linear trend, while a time series integrated of order 2 would have a quadratic trend.  Maintaining a quadratic trend over any significant time is unlikely without the measured thing shrinking to nothing or actually exploding.

There are two somewhat easy ways to estimate the fractional integration order: Hurst's exponent and by visual inspection of the ACF.  

Hurst's exponent was developed to describe the variability of flooding in the Nile river delta, but it is an interesting measurement in its own right.  It represents a measure of the short and long term dependency of a time series on itself.  It's like a more general autocorrelation and produces sequences that reflect certain phenomena very well, but are notoriously difficult to calculate.  Even estimating the Hurst exponent isn't a closed problem, with several reasonable approaches in the literature.

Hurst's exponent takes subsets of the time series of a particular length, say, $n$ and then estimates the "scaled range" in the data subsets of length $n$.  The scaled range is the difference of the max and min values in the subset, divided by the standard deviation in the subset.  As $n$ is varied, this produces a relationship between the size of the subset window and the scaled range of the time series.

When this result is plotted on a log-log plot the points tend to follow a line (power law) and the slope of this line is an estimate of the Hurst exponent, a fascinating number that characterizes a long term relationship in the time series.

When visualizing the ACF, note the rate at which the autocorrelation decays. The degree of fractional integration is proportional to:

$$ k^{2d-1}$$

Where $k$ is the lag of the autocorrelation and $d$ is the integration dimension.

# Mutual Information

Okay, I'm seriously running out of time on this thing, but I wanna try to get to a cool final point.  I'm not even sure if there's gonna be enough time to cover all this in the brown bag talk.

Real quick: KL divergence is a divergence (not a distance - in that it is 0 when two things are equal and is always > 0 when two things aren't equal, but doesn't obey the triangle inequality and isn't symmetric) that can be used to characterize the difference between two distributions

Mutual information can be thought of as the amount of information knowing the value in one distribution gives us about the value in another distribution.  The amazing and powerful thing about mutual information is that it isn't restricted to a linear relationship, or any particular relationship type - it just indicates how much information is shared.  In fact, mutual information is invariant under arbitrary deterministic nonlinear transforms (meaning that the mutual information between $a$ and $b$ is the same as it is between $a$ and $F(b)$ where $F$ is an arbitrary transform).

The mutual information can be characterized in a number of ways, but my favorite is:

$$ I(X,Z) = D_{KL}(P_{XZ}||P_X \otimes P_Z) $$

Which says the the mutual information between two distributions (two data series) is the KL divergence between the joint distribution and the product of the marginal distributions (assumes independence).

One of the reasons I used some clunky notation around stationarity and correlations (e.g. the conditional value of mean given some condition) is that's similar to how mutual information is thought of.  

The Achilles heel of mutual information is that it can only really be estimated (for discrete, sampled series) for categorical or discrete data.  Continuous values can't be estimated because the mutual information is the maximum possible relationship value between two sets across the set of all nonlinear transforms.

But a few years ago a paper came out called [MINE](https://arxiv.org/pdf/1801.04062) which leverages the power of neural networks as universal function approximators to actually perform the search over all nonlinear transforms and estimate the mutual information betweent two continuously-valued discrete time series.

If anyone wants to push the state of the art in time series analysis, I'd suggest they work on reframing many of the fragile tests and methods currently used in time series analysis (requiring stationarity, manual transforms, etc.) and convert them to a pure mutual information system.



### MINE Example code

```python
# Stolen from: https://github.com/MasanoriYamada/Mine_pytorch/blob/master/mine.ipynb

# data
var = 0.2
def func(x):
    return x

def gen_x():
    return np.sign(np.random.normal(0.,1.,[data_size,1]))

def gen_y(x):
    return func(x)+np.random.normal(0.,np.sqrt(var),[data_size,1])
```

```python
data_size = 1000000
x=gen_x()
y=gen_y(x)
p_y_x=np.exp(-(y-x)**2/(2*var))
p_y_x_minus=np.exp(-(y+1)**2/(2*var))
p_y_x_plus=np.exp(-(y-1)**2/(2*var))
mi=np.average(np.log(p_y_x/(0.5*p_y_x_minus+0.5*p_y_x_plus)))
mi
```

```python
H=10
n_epoch = 500
data_size = 20000

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1, H)
        self.fc2 = nn.Linear(1, H)
        self.fc3 = nn.Linear(H, 1)

    def forward(self, x, y):
        h1 = F.relu(self.fc1(x)+self.fc2(y))
        h2 = self.fc3(h1)
        return h2    

model = Net()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
plot_loss = []
for epoch in tqdm(range(n_epoch)):
    x_sample=gen_x()
    y_sample=gen_y(x_sample)
    y_shuffle=np.random.permutation(y_sample)
    
    x_sample = Variable(torch.from_numpy(x_sample).type(torch.FloatTensor), requires_grad = True)
    y_sample = Variable(torch.from_numpy(y_sample).type(torch.FloatTensor), requires_grad = True)
    y_shuffle = Variable(torch.from_numpy(y_shuffle).type(torch.FloatTensor), requires_grad = True)    
    
    pred_xy = model(x_sample, y_sample)
    pred_x_y = model(x_sample, y_shuffle)

    ret = torch.mean(pred_xy) - torch.log(torch.mean(torch.exp(pred_x_y)))
    loss = - ret  # maximize
    plot_loss.append(loss.data.numpy())
    model.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plot_x = np.arange(len(plot_loss))
plot_y = np.array(plot_loss).reshape(-1,)
hv.Curve((plot_x, -plot_y)) * hv.Curve((plot_x, mi))
```

## Models

Now that the time series has been fully characterized - the distributions and parameters have been estimated, it's been made stationary, and any suspected component signals have been isolated and characterized, now models can be chosen to represent the time series.  Some popular models for time series models include

* Markov chain
* ARIMA(X) / ARFIMA(X)
* State space models

## Cointegration

**Side note about an important concept**

For nonstationary data there are a still approaches which directly estimate the relationship between multiple variables.  Cointegration is probably the best known among these.

Cointegration tries to get around the nonstationarity of data (which can create a spurious correlation) by instead looking at the difference between the two time series.  More specifically a parameter, $\beta$ is used in an attempt to fit one time series to the other in such a way that the difference between the time series is stationary.

The argument is that if the difference between the two time series remains stationary (in this case usually a more limited measure of mean and variance, not including autocorrelation) that there is a relationship between the two time series that goes beyond their nonstationarity.  A bit more rigorously, it's assumed that the processes being compared are constructed of a constant, a stationary process, and a (cumulatively summed) stochastic process. ## Synthesis

Distinct from a forecast, synthesis is about building a stochastic process that will mimic the behavior of the time series.  The output will be a non-deterministic "realization" of a possible future trace.  Many runs will create a boundary of sorts for the stochastic evolution of the process.  The better the stochastic process mimics the time series, the more of the idiosyncratic character of the time series will be captured by the simulation.

Synthesis is useful in creating scenarios, understanding variability, and generating synthetic (but characteristically consistent) time series to power simulations.  It will combine all the techniques presented so far.

### Bonus Fun - Hilbert Transform

Sometimes a time series will have periodic components that change in frequency over time.  Once the stationary periodic component is extracted it becomes important to be able to characterize the nature of the periodic nonstationarity.  A Hilbert transform is extremely useful for this.

A Hilbert transform adds the complex conjugate to the original signal, providing an interpretation of the signal with instantaneous phase and amplitude.  Roughly, it can be thought of as taking an FFT of the data, but then only paying attention to half of the FFT.  FFTs wrap the time series around the origin in the complex plane, or can equivalently be thought of as a spinning vector, centered at the origin of the complex plane, that can change length as it traces out the time series.  The projection of this complex plane vector onto the real line is the original time series.

If the complex-valued spinning vector is ever going to map back to the real line it must have a counterpart - a counter-spinning vector with an equal and opposite complex component.  This is one interpretation of the negative frequencies produced in an FFT.  If one of the two spinning vectors are eliminated, there will only be a single spinning vector left, the real component of which traces out the time series on the real axis.  The angle of this vector is the "instantaneous phase" while the length of the vector is the "instantaneous magnitude."

Taking the FFT of a time series and setting all the negative frequency components to 0 is a more roundabout, but equivalent way to produce an analytic signal.

The instantaneous magnitude will form an envelope around a time series (especially if its mean stationary - if it's not mean stationary the envelope will tend to follow the line and the phase will stay locked near 0).  This envelope can be thought of as the time-local range of the function, and the frequency can be useful in extracting information about periodic time series.  For a basic FFT analysis low frequency and high frequency components will be sort-of averaged and sort-of canceled out by other frequencies.  Hilbert transforms allow magnitude and frequency to vary with time, making it a powerful tool similar to wavelet analysis to evaluate frequency- and magnitude-varying time series.

```python
#t0 = np.
y0 = np.cos(np.linspace(0,8*np.pi, 1000)**1.7) + np.cos(np.linspace(0,9*np.pi, 1000)**1.7)
y0 = y0 * np.linspace(1,2,1000)

analytic_signal = signal.hilbert(y0)
amplitude_envelope = np.abs(analytic_signal)
instantaneous_phase = np.unwrap(np.angle(analytic_signal))
instantaneous_frequency = (np.diff(instantaneous_phase) /
                            (2.0*np.pi))

plt.subplot(2,1,1)
plt.plot(y0, linewidth=2)
plt.plot(amplitude_envelope)
plt.plot(-amplitude_envelope)
plt.subplot(2,1,2)
plt.plot(instantaneous_frequency)
plt.gcf().set_size_inches(12,12)
```
