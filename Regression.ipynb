# Regression
Planning to use two predictors (time, closing price of eBay stock) to model the closing stock price of the Amazon stock. 
  
I am going to follow these steps:  
1. Explore the relationship between predictors and variable  
2. Fit one model for each predictor
3. Analyze the distribution of residuals to analyze the fit  


## House keeping
Importing  libraries and loading data

```python
# Importing useful libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import scipy.stats as stats
import statsmodels.api as sm
```

```python
# Reading in data
data = pd.read_csv('data/prices.csv')

# Getting the amazon stock price
amzn = data.loc[data['symbol'] == 'AMZN'] 
# Getting the two predictors
# Predictor 1. eBay stock
ebay = data.loc[data['symbol'] == 'EBAY'] 
finaldata = amzn.merge(ebay, on='date')
# Predictor 2. Time
finaldata['t'] = np.arange(len(finaldata))
```

## Exploratory data analysis

### Checking

```python
features = ['close_x', 'close_y', 't']
plt.matshow(finaldata[features].corr())
plt.xticks([0,1,2],features)
plt.yticks([0,1,2],features)
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(finaldata[features]);
```

```python

X = finaldata["t"]
y = finaldata["close_x"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```

```python
plt.plot(np.arange(len(model.resid)), model.resid)
plt.show()
```

```python
n, x, _ = plt.hist(model.resid, density=True)
density = stats.gaussian_kde(model.resid)
plt.plot(x, density(x))
plt.xlabel("Residual value")
plt.ylabel("Probability density")
plt.show()
```

```python
testStatistic, pValue = stats.shapiro(model.resid)
if pValue < 0.05:
    print("Residuals are not normally distributed")
else:
    print("Residuals are normally distributed")
```
