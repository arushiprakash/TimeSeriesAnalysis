```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/prices.csv')
```

```python
data.columns
```

```python
stocks_of_interest = {
    'AMZN' : 'Amazon',
    #'SFIX' : 'Stitch Fix',
    #'SHOP' : 'Shopify',
    #'BABA' : 'Alibaba',
    'EBAY' : 'eBay',
    #'W'    : 'Wayfair',
    #'ETSY' : 'Etsy'
}
```

## Visualizing the stock prices
The most important thing to do is visualize the time series data. Here, we plot the graphs simply

```python
for key in stocks_of_interest.keys():
    sub_ = data.loc[data['symbol'] == key] 
    plt.plot(sub_['date'], sub_['close'], label=key)
    plt.xticks([])
    plt.title("Closing stock prices over time")
    plt.ylabel("Stock Price ($)")
    plt.xlabel("Trading Day")
    plt.legend()
    plt.show()
```
